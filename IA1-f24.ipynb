```python
def quicksort(arr):

    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quicksort(left) + middle + quicksort(right)

# Example usage
if __name__ == "__main__":
    array = [6.3, 4.2, 1.5, 2.7]
    print("Original array:", array)
    sorted_array = quicksort(array)
    print("Sorted array:", sorted_array)
```

```python
import random

def quicksort(arr):

    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        return quicksort(left) + middle + quicksort(right)

def generate_random_list(n):

    return [random.uniform(0, 1) for _ in range(n)]

# Test cases for n = 0, 1, 10, 500
test_values = [0, 1, 10, 500]

for n in test_values:
    # Generate the random list
    random_list = generate_random_list(n)
    print(f"\nOriginal list for n={n}:", random_list)

    # Sort the list using quicksort
    sorted_list = quicksort(random_list)
    print(f"Sorted list for n={n}:", sorted_list)
```
